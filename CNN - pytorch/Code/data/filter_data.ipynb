```python
from pycocotools.coco import COCO
import json

dataDir = r'C:\Users\thijs\University\Year4\Thesis\Pract\Data'
annFile = '{}/annotations/trial_val.json'.format(dataDir)
coco = COCO(annFile)
```

```python
allCatIDs = coco.getCatIds()
print(f"N of cats: {len(allCatIDs)}")

i = 0
annoN = []
catNames = []
# Get all available IDS & Go through them one by one
for catID in coco.getCatIds():
    # Get the category info
    cat = coco.loadCats(catID)[0]
    catName = cat['name']
    annoIDs = coco.getAnnIds(catIds=catID)
    print(f"{catID} - {catName} - {len(annoIDs)}")
    annoN.append(len(annoIDs))
    catNames.append(catName)
    i += len(annoIDs)
print(f"Total annotations: {i}")
```

```python
personID = 1
```

```python
newSort = sorted(annoN, reverse=True)

largestCats = []
largestCatsIDs = []
# Shows the largest (most frequently ocurring) labels:
for N_anno in newSort[:10]:
    ID = annoN.index(N_anno)
    category = catNames[ID]
    print(ID+1, '-', category, N_anno)
    largestCatsIDs.append(ID+1)
    largestCats.append(category)
print(largestCatsIDs)
largestCats = coco.loadCats(largestCatsIDs)
print(largestCats)
```

```python
selectedCatIDs = largestCatsIDs[:5]
selectedCatIDs = []
print(f"IDs to keep: {selectedCatIDs}")

annoIDs = coco.getAnnIds(catIds=selectedCatIDs)
annotations = coco.loadAnns(annoIDs)
keep_anno_ID = [anno['id'] for anno in annotations]
keep_anno = coco.loadAnns(keep_anno_ID)
print(f"First 10 annotations to keep: {coco.loadAnns(keep_anno_ID[:10])}")
print(annotations[0].keys())
```

```python
with open(annFile, 'r') as f:
    ann_data = json.load(f)      

new_ann_data = {
    'info': ann_data['info'],
    'licenses': ann_data['licenses'],
    'images': ann_data['images'],
    'annotations': [anno for anno in keep_anno if anno['category_id'] in selectedCatIDs],
    'categories': [cat for cat in ann_data['categories'] if cat['id'] in selectedCatIDs] 
}

print(new_ann_data['categories'])

# Save the modified annotation file to a new JSON file
new_ann_file = r'C:\Users\thijs\University\Year4\Thesis\Pract\Data\annotations\selected_classes.json'
with open(new_ann_file, 'w') as f:
    json.dump(new_ann_data, f)
```
